```python
from urllib import request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://fishing-you.jp/item_list"
response = request.urlopen(url)
bs = BeautifulSoup(response)
a_all = bs.find_all("div", class_="category_area_body")[0].find_all("a")

#カテゴリのURLリスト作成
categories = []
for a in a_all:
    a_href = a.attrs["href"]
    if len(a_href) < 16:
        categories.append("https://fishing-you.jp" + a_href)
categories = ['https://fishing-you.jp/item_list/001/']
```

```python
#名前と値段のリスト
item_name_lis = []
price_lis = []
i=0
#1カテゴリずつ取得していく
for url in categories:
    response = request.urlopen(url)
    bs = BeautifulSoup(response)
    print(url)
    
    #ページ番号に「次のページへ」が存在するかどうか
    next_page_text = bs.find_all("div", id="pager_top")[0].find_all("a")[-1].text
    while next_page_text == "次のページへ":
        i += 1
        #現在のページのアイテムの情報を取得
        items = bs.find_all("div", class_="item-box")
        for item in items:    
            item_name = item.find("p", class_="pn").find("a").text
            price = item.find("p", class_="sp").find("strong").text
            item_name_lis.append(item_name)
            price_lis.append(price)
        
        #「次のページへ」のURL取得
        next_page_href = bs.find_all("div", id="pager_top")[0].find_all("a")[-1].attrs["href"]
        url = "https://fishing-you.jp" + next_page_href
        print(i)
        response = request.urlopen(url)
        bs = BeautifulSoup(response)
        next_page_text = bs.find_all("div", id="pager_top")[0].find_all("a")[-1].text
        
df = pd.DataFrame(list(zip(item_name_lis, price_lis)), columns = ['Name','Price'])
df
```
